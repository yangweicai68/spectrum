```python
##加载包
import numpy as np
import pandas as pd
##import os

##读取叶面积指数
df=pd.read_excel(r'D:\1博士2\2019试验数据\高光谱\植被指数叶面积相关性-平均值.xlsx')
df[:5]
```

```python
##散点图
%pylab inline
result=pd.scatter_matrix(df,diagonal='kde',color='k',alpha=0.3,figsize=(10,10))     ####diagonal='hist',显示直方图，=kde显示和密度估计
```

```python
##相关系数矩阵图
import matplotlib.pyplot as plt
import seaborn as sns

corr=df.corr()
corr=abs(corr)

fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(figsize=(10,8))    ###定义一个子图

ax=sns.heatmap(corr,vmax=1,vmin=0,annot=True,annot_kws={'size':13,'weight':'bold'},linewidths=0.05)

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.show()
fig.savefig(r'D:\1博士2\2019试验数据\高光谱\叶面积相关系数总-不覆膜平均值.png') 
```
